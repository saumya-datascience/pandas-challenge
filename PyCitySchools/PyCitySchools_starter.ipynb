```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()

#I think there is no reason use groupby, because in code are not count some values per groups.

#conditions = [
 #   ((df["Balance"]>0) & (df['Value']>0) & (df['L_Date']<=df['c_date'])),
  #  ((df["Balance"]==0) & (df['Value']==0) & (df['L_Date']<=df['c_date']))]
#choices = ['Good', 'Bad']
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
school_count=school_data_complete["School ID"].value_counts()
total_school=len(school_count)
print(total_school)

#Calculate the total number of students
total_students=school_data_complete["Student ID"].count()
print(total_students)

#Calculate the total budget
tot_budget=school_data["budget"].sum()
print(tot_budget)

#Calculate the average math score
avg_maths=school_data_complete["math_score"].mean()
print(avg_maths)

#avearge reading score
avg_reading=school_data_complete["reading_score"].mean()
print(avg_reading)

#calculate the percentage of students with a passing math score(70 or greater)
per_passmaths=(len(school_data_complete[school_data_complete.math_score >=70])/total_students)*100
print(per_passmaths)

#calculate the percentage of students with  passing reading score(70 or greater)
per_passread=(len(school_data_complete[school_data_complete.reading_score>=70])/total_students)*100
print(per_passread)

#Calculate the percentage of students who passed math and reading (% Overall Passing)
per_read_math=len(school_data_complete[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70)])
percent_read_math=per_read_math/total_students*100
percent_read_math

district_df=pd.DataFrame({"Total Schools":total_school,
                         "Total Students":total_students,
                         "Total Budget":tot_budget,
                         "Average Math Score":avg_maths,
                         "Average Reading Score":avg_reading,
                         "% Passing Math":per_passmaths,
                         "% Passing Reading":per_passread,
                         "% Overall Passing":percent_read_math}, index=[0])
district_df
```

* Create an overview table that summarizes key metrics about each school, including:

```python
#school name

school_info_df=school_data_complete.groupby("school_name")
name=school_info_df["school_name"].unique()
#School Type
name_type=school_info_df["type"].unique()
#total Students
total_student=school_info_df["Student ID"].count()
#total_School Budget
tot_budget=school_info_df["budget"].mean()
#Per Student Budget
per_school_counts = school_data_complete["school_name"].value_counts()
per_stu_budget=tot_budget/per_school_counts

#print(average_math)
average_math=school_info_df["math_score"].mean()
#Average reading score
average_reading=school_info_df["reading_score"].mean()
#print(pass_percent_maths)
school_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
pass_percent_maths=school_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
#% passsing reading
school_passing_reading=school_data_complete[(school_data_complete["reading_score"]>=70)]
per_school_reading=school_passing_reading.groupby(["school_name"]).count()["student_name"]/per_school_counts*100

#% overall passing
pass_reading_math=school_data_complete[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70)]
Overall_Passing= pass_reading_math.groupby("school_name").count()["student_name"]/per_school_counts*100

school_summary_df=pd.DataFrame({"School Type":name_type,
                      "Total School Budget":tot_budget,
                      "Total Students":total_student,
                      "Per Student Budget":per_stu_budget,
                      "Average Math Score":average_math,
                      "Average Reading Score":average_reading,
                      "% Passing Math":pass_percent_maths,
                      "% Passing Reading":per_school_reading,
                      "% Overall Passing":Overall_Passing})

school_summary_df.head()

#data=[name_type,tot_budget,total_student,per_stu_budget,average_math,average_reading,pass_percent_maths,per_school_reading,Overall_Passing]

#headers=["School Type","Total School Budget","Total Students","Per Student Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]
#df3 = pd.concat(data,axis=1,keys=headers)
#df3







```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
nineth_df = school_data_complete.loc[school_data_complete["grade"]=="9th",["school_name","math_score"]]
group_nine=nineth_df.groupby("school_name")["math_score"].mean()                              
group_nine.head()
tenth_df = school_data_complete.loc[school_data_complete["grade"]=="10th",["school_name","math_score"]]
group_ten=tenth_df.groupby("school_name")["math_score"].mean()                              
group_ten.head()
eleven_df = school_data_complete.loc[school_data_complete["grade"]=="11th",["school_name","math_score"]]
group_eleven=eleven_df.groupby("school_name")["math_score"].mean()                              
group_eleven.head()
twelve_df = school_data_complete.loc[school_data_complete["grade"]=="12th",["school_name","math_score"]]
group_twelve=twelve_df.groupby("school_name")["math_score"].mean()                              
group_twelve.head()
d=(group_nine,group_ten,group_eleven,group_twelve)
headers=["9th","10th","11th","12th"]
math=pd.concat(d,axis=1,keys=headers)
math.head()
#school_data_complete.groupby(['school_name','grade' as_index=False])['math_score'].mean()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#min_value=school_summary_df["Per Student Budget"].min()

hp_bins = [0,585, 630, 645,680]
hp_labels = ["<585", "$585-630", "$630-645","$645-680"]

# Bin the Horsepower column
# cut() returns a Pandas Series containing each of the binned column's values translated into their corresponding bins

school_summary_df["spending_ranges"]=pd.cut(school_summary_df["Per Student Budget"], hp_bins, labels=hp_labels)

spending_ranges=school_summary_df.groupby(["spending_ranges"])["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"].mean()
spending_ranges.head()
```
